### 1) **Algorithme de soustraction des frames par indication du décalage temporel (la variable timeLapse)**

```python

import cv2
import numpy as np
import matplotlib.pyplot as plt


def FrameSubstraction(file_path,timeLapse):

    

    video = cv2.VideoCapture(file_path)
    liste=[]
    i=0

    # running the loop
    while(video.isOpened()):

        # extracting the frames
        success, current = video.read()
        
        if  success:   
            image = cv2.cvtColor(current, cv2.COLOR_BGR2RGB)
            gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

            if i<timeLapse:             # We cannot treat frame < timeLapse, so we just save each one in a list
                liste.append(gray)

            else:
                
                result=cv2.absdiff(liste[i-timeLapse],gray)
                _, result = cv2.threshold(result,100,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
                cv2.imshow('Real scene',liste[i-timeLapse])
                cv2.imshow('FrameSubstraction with '+str(timeLapse)+' frames',result)
                
                liste.append(gray)
                
            
            i+=1
            
            key = cv2.waitKey(50)
            if key == ord("q"):
                print( "Vidéo arretée")
                break  
        else:    
            print( "FIN DE LA VIDÉO")
            #end of the video
            video.release()
            cv2.destroyAllWindows()
            
            break
        
        
file_path = "/Users/khelifibilel/TP_VIDEO/video_1.mp4"  

timeLapse=20
FrameSubstraction(file_path,timeLapse)
```

### 2) THREE-FRAME DIFFERENCE

(à modifier)
Three-frame difference method is an improved algorithm based on two-frame difference. It puts three adjacent frames as a group, subtracts both two adjacent frames and lets two differential results do the logical AND operation. 

```python

import cv2
import numpy as np
import matplotlib.pyplot as plt


def three_fr_Differencing(file_path,timeLapse):

    video = cv2.VideoCapture(file_path)
    liste=[]
    i=0
    kernel = np.ones((7,7),np.uint8)   #for opening and closing operations

    
    while(video.isOpened()):

        # extracting the frames
        success, current = video.read()
        
        if  success:   
            image = cv2.cvtColor(current, cv2.COLOR_BGR2RGB)
            gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
            
            #otsu_threshold, gray = cv2.threshold(gray,100,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)

            
            if i < 2*timeLapse:
                liste.append(gray)

            else:
                
                
                print('Previous frame:',i-2*timeLapse+1)
                print('Current frame:',i-timeLapse)
                print('Next frame:',i-1)
                print('-'*15)
                
                
                result1=cv2.absdiff(liste[i-2*timeLapse+1],liste[i-timeLapse])   # substract frame(t)-frame(t+timeLapse)
                result2=cv2.absdiff(liste[i-1],liste[i-timeLapse])              # substract frame(t)-frame(t-timeLapse)
                
                result = cv2.bitwise_and(result1, result2)           #logical AND between (t,t-timeLapse) and (t,t+timeLapse)
                
                cv2.imshow('Real scene',liste[i-timeLapse])
                
                _, result = cv2.threshold(result,100,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
                
                
                #result = cv2.morphologyEx(result, cv2.MORPH_OPEN, kernel)
                result = cv2.morphologyEx(result, cv2.MORPH_CLOSE, kernel)
                

                #blur = cv2.GaussianBlur(result,(3,3),2)
                #median = cv2.medianBlur(result,3)

                cv2.imshow('Three_fr_Differencing: '+str(timeLapse-1)+' frames',result)
                
                liste.append(gray)
                
            
            i+=1
            
            key = cv2.waitKey(20)
            if key == ord("q"):
                break  
        else:    
            #end of the video
            print( "FIN DE LA VIDÉO")
            video.release()
            cv2.destroyAllWindows()
            
            break
        
        
file_path = "/Users/khelifibilel/TP_VIDEO/video_1.mp4"  

timeLapse=2

three_fr_Differencing(file_path,timeLapse+1)
```

## 3) Adaptive Background Substruction

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt


def Adap_Bg_Substruction(file_path,timeLapse):

    

    video = cv2.VideoCapture(file_path)
    liste=[]
    i=0
    alpha=0.6

    while(video.isOpened()):

        # extracting the frames
        success, current = video.read()
        
        if  success:   
            image = cv2.cvtColor(current, cv2.COLOR_BGR2RGB)
            gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
           
            if i<timeLapse:
                liste.append(gray)

            else:
                
                
                result = cv2.addWeighted(gray, alpha, liste[i-timeLapse], 1-alpha, 0.0)
                _, result = cv2.threshold(result,100,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
                
                
                
                cv2.imshow('Real scene',gray)
                

                cv2.imshow('FrameSubstraction with '+str(timeLapse)+' frames for alpha= {:.1f}'.format(alpha)
                           ,result)
                
                liste.append(result)
                
            
            i+=1
            
            key = cv2.waitKey(30)
            if key == ord("q"):
                print( "Vidéo arretée")
                break  
        else:    
            print( "FIN DE LA VIDÉO")
            #end of the video
            video.release()
            cv2.destroyAllWindows()
            
            break
        
        
file_path = "/Users/khelifibilel/TP_VIDEO/video_1.mp4"  

timeLapse=2
Adap_Bg_Substruction(file_path,timeLapse)
```

## 4) Background estimation and substraction

BS calcule le masque de premier plan en effectuant une soustraction entre l'image courante du modèle BG, contenant la partie statique de la scène.
La modélisation du fond est faite par 2 étapes:

1- Initialisation du fond 
2- Mise à jour du régulier du BG

#### 1. MOG est une méthode paramétrique qui suppose que la distribution d'intensité d'une image est normal, les méthodes MOG estime une probabilité d'intensité pour chaque pixel est fait la classification sur cette base.
MOG est très sensible au mouvement de la caméra et au changement continue du background.

L'amélioration apportée par MOG2 par rapport MOG est qu'il sélectione automatiquement le nombre de kernels gaussiens utilisés pour chaque pixel, à la place du nombre constant pour MOG => MOG2 offre une meilleure adaptabilité aux conditions d'illumination qui changent dans les différentes scènes [1]. 

#### 2. KNN est une méthode non paramétrique qui estime l'intensité de chaque pixel de manière indépendante, elle se base sur un calcul de similarité entre les pixels pour distinguer entre background et foreground.
On stocke tous les cas disponibles, puis classer les nouveaux échantillons en fonction de la mesure de similarité. Cela peut être un inconvénient pour des données de taille importante.
KNN est plus robuste au mouvement de la caméra et au changement du BG mais trop lent dans la prise de décision.



### 4.1) KNN

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

backSub = cv2.createBackgroundSubtractorKNN(detectShadows=True)
capture = cv2.VideoCapture("/Users/khelifibilel/TP_VIDEO/video_1.mp4")
if not capture.isOpened():
    print('Unable to open ')
    exit(0)
while capture.isOpened():
    ret, frame = capture.read()
    if frame is None:
        break
    
    frame = cv2.resize(frame, (960, 540))
    fgMask = backSub.apply(frame)
    cv2.rectangle(frame, (10, 2), (100,20), (255,255,255), -1)
    cv2.putText(frame, str(capture.get(cv2.CAP_PROP_POS_FRAMES)), (15, 15),
               cv2.FONT_HERSHEY_SIMPLEX, 0.5 , (0,0,0))
    
    
    cv2.imshow('REAL SCENE',frame)
    
    a, result = cv2.threshold(fgMask,100,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
    cv2.imshow('KNN',result)
    
    keyboard = cv2.waitKey(30)
    if keyboard == 'q':
        break
```
